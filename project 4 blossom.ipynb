## Building dataset

```python
records = []
for result in results:
    property_name = result.find('a')['href'][1:-20]
    beds = result.find('li', attrs={'class':'bed'}).find('span').text
    showers = result.find('li', attrs={'class':'shower'}).find('span').text
#     garages = result.find('li', attrs={'class':'garage'}).find('span').text
#     area = result.find('li', attrs={'class':'area'}).find('span').text
    description = result.find('img')['alt']
    price = result.find('p', attrs = {'span class':None}).text[9:-9]
    currency = result.find('p', attrs = {'span class':None}).text[6:-15]
    rent_period = result.find('p').text[16:-1]
    url = 'www.meqasa.com' + result.find('a')['href']
    address = result.contents[5].find('a')['href'][22:-13]
    time_posted = result.find('p', attrs={'class':'wsnr'}).text
    records.append((property_name, beds, showers, {}, {}, description, price, currency, rent_period, url, address, time_posted))
```
